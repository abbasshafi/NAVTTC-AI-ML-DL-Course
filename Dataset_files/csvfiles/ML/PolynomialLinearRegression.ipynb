```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing the DataSet
dataset=pd.read_csv('D:\AI and DL Course\MachineLearningCourse\Machine Learning A-Z (Codes and Datasets)\Part 2 - Regression\Section 6 - Polynomial Regression\Python\\Position_Salaries.csv')

x=dataset.iloc[: ,1:-1].values
y=dataset.iloc[: , -1].values
```

```python
print(x)
```

```python
print(y)
```

```python
#Training the Linear Regression Model on the whole datset
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

```python
#Training The Polynomial regression on the whole dataset
from sklearn.preprocessing import PolynomialFeatures
polyreg=PolynomialFeatures(degree=4)
x_poly=polyreg.fit_transform(x)
lin_reg2=LinearRegression()
lin_reg2.fit(x_poly,y)
```

```python
#Visualizing the Linear Regresiion results
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='yellow')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
##Visualizing the Polynomial Regresiion results
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg2.predict(x_poly),color='blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Predicting a new result with Linear Result ,Predciting single Observation
lin_reg.predict([[6.5]])
```

```python
#Predicting a new result with Polynomial Result ,Predciting single Observation\
lin_reg2.predict(polyreg.fit_transform([[6.5]]))
```
